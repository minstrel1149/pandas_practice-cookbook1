```python
years = 2016, 2017, 2018
stocks = [pd.read_csv(p1 / f'stocks_{year}.csv', index_col='Symbol') for year in years]
stocks_16, stocks_17, stocks_18 = stocks
stocks_16.head()
```

```python
pd.concat([stocks_16, stocks_17, stocks_18], keys=[2016, 2017, 2018], names=['Year', 'Symbol'], axis=0)
```

```python
stock_concat = pd.concat([stocks_16, stocks_17, stocks_18], keys=[2016, 2017, 2018], names=['Year', None], axis=1)
stock_concat
```

```python
(stocks_16
 .join(stocks_17, lsuffix='_2016', rsuffix='_2017', how='outer')
 .join(stocks_18.add_suffix('_2018'), how='outer')
)
```

```python
other = [stocks_17.add_suffix('_17'), stocks_18.add_suffix('_18')]
# join에 리스트를 집어넣는 것도 가능
(stocks_16
 .add_suffix('_16')
 .join(other, how='outer')
 .sort_index()
)
```

```python
(stock_concat
 .rename_axis([None, None], axis=1)
 .swaplevel(0, 1, axis=1)
 .pipe(lambda df: df.set_axis(df.columns.to_flat_index(), axis=1))
 .rename(lambda cols: '_'.join([str(col) for col in cols]), axis=1)
 .sort_index()
)
```

```python
(stocks_16
 .merge(stocks_17, left_index=True, right_index=True, how='outer', suffixes=['_16', '_17'])
 .merge(stocks_18.add_suffix('_18'), left_index=True, right_index=True, how='outer')
)
```
