```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
p1 = Path.cwd() / 'back_data'
movie = pd.read_csv(p1 / 'movie.csv')
movie.head()
```

```python
mov_act_dir = (movie
               .loc[:, [col for col in movie.columns if '_name' in col and ('actor' in col or 'director' in col)]]
               .reindex(['actor_1_name', 'actor_2_name', 'actor_3_name', 'director_name'], axis=1)
)
mov_act_dir.head()
```

```python
def shorten(col):
    return str(col).replace('facebook_likes', 'fb').replace('_for_reviews', '')
movie = movie.rename(shorten, axis=1)
movie.head()
```

```python
movie.select_dtypes(include='int').head(), movie.select_dtypes(include='number').head()
```

```python
college = pd.read_csv(p1 / 'college.csv', index_col='INSTNM')
college.head()
```

```python
college_ugds = college.filter(like='UGDS_')
college_ugds.head()
```

```python
name = 'Northwest-Shoals Community College'
college_ugds.loc[name, :]
```

```python
# pandas는 bankers rounding 연산 활용 → UGDS_BLACK이 예시
college_ugds.loc[name, :].round(2)
```

```python
(college_ugds.loc[name, :] + 0.00001).round(2)
```

```python
# Nan은 자기 자신과 같지 않음, Nan을 대상으로 한 모든 비교는 !=를 제외하고는 모두 False
np.nan == np.nan
```

```python
(college_ugds == college_ugds).all()
```

```python
(college_ugds.eq(college_ugds)).all()
```

```python
# assert_frame_equal은 같지 않을 경우 Assertion Error 발생
pd.testing.assert_frame_equal(college_ugds, college_ugds, check_dtype=False) == None
```
