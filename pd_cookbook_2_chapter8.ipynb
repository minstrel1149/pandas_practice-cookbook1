```python
def remove_binary_cols(df):
    binary_only = df.nunique().eq(2)
    cols = binary_only.loc[binary_only].index.tolist()
    return df.drop(cols, axis=1)

def select_rows_with_max_cols(df):
    max_cols = df.idxmax()
    unique = max_cols.unique()
    return df.loc[unique]
```

```python
(college
 .select_dtypes('number')
 .pipe(remove_binary_cols)
 .pipe(select_rows_with_max_cols)
 .head()
 .style.highlight_max()
)
```

```python
col_n = (college
 .select_dtypes('number')
 .pipe(remove_binary_cols)
)
col_n.head()
```

```python
col_n.max().head()
```

```python
has_row_max2 = (col_n
 .eq(col_n.max())
 .cumsum().cumsum()
 .eq(1)
 .any(axis=1)
)
has_row_max2
```

```python
idxmax_cols = has_row_max2.loc[has_row_max2].index
idxmax_cols
```

```python
col_n.loc[idxmax_cols, :].style.highlight_max()
```

```python
set(col_n.idxmax().unique()) == set(idxmax_cols)
```
