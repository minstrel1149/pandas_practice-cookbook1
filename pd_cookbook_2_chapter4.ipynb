```python
movie = pd.read_csv(p1 / 'movie.csv')
movie2 = movie.loc[:, ['movie_title', 'imdb_score', 'budget']]
movie2.head()
```

```python
# sort_values를 활용할 시 kind='mergsort'를 전달하면 동일한 출력
(movie2
 .nlargest(100, 'imdb_score')
 .nsmallest(5, 'budget')
)
```

```python
(movie2
 .query("imdb_score >= 8.4")
 .nsmallest(5, 'budget')
 .sort_values(['budget', 'imdb_score'], ascending=[True, False])
)
```

```python
movie3 = movie.loc[:, ['movie_title', 'title_year', 'imdb_score']].dropna(how='any')
movie3.head()
```

```python
(movie3
 .sort_values(['title_year', 'imdb_score'], ascending=[False, False])
 .drop_duplicates(subset='title_year', keep='first')
)
```

```python
(movie3
 .groupby('title_year', as_index=False)
 .apply(lambda df: df.sort_values('imdb_score', ascending=False).head(1))
 .droplevel(0)
 .sort_values('title_year', ascending=False)
)
```
